### IMPORTING LIBRARIES

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import mean_absolute_error as MAE
from sklearn.metrics import mean_squared_error as MSE
```

### LOADING DATASET PROVIDED

```python
data= "http://bit.ly/w-data"
df= pd.read_csv(data)
df.head(10)
```

### DATA VISUALISATION 

```python
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Score(Fig 1)')  
plt.xlabel('Hours Spent Studying')  
plt.ylabel('Percentage Scored')
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title('Hours vs Score(Fig 2)')
sns.heatmap(df.corr(),annot=True,cmap="PuBuGn")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.regplot(x=df['Hours'],y=df['Scores'],data=df,color='red')
plt.title('Study Hours vs  Scores')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```

### TRAINING THE DATASET FOR ML MODELLING

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
```

##### BUILDING OUR LOGISTIC REGRESSION MODEL

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
line = lr.intercept_+lr.coef_*X
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(df['Hours'],df['Scores'],s=200,color='blue')
plt.plot(df['Hours'],line,'b')
plt.xlabel(xlabel='Hours',fontsize=20)
plt.ylabel(ylabel='Scores',fontsize=20)
plt.show()
```

### Predicting our Results and Comparing it With Actual Dataset

```python
print(X_test) 
y_predicted = lr.predict(X_test)
```

```python
datafr = pd.DataFrame({'Actual': y_test, 'Predicted': y_predicted})
datafr
```

```python
hours = np.array([9.25])
hours = hours.reshape(-1,1)
own_pred = lr.predict(hours)
print(f"No of Hours = {hours}")
print(f"Predicted Score = {own_pred[0]}")
```

### Final Evaluation of Model

```python
from sklearn import metrics 
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_predicted)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_predicted))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_predicted)))
print('R-2:', metrics.r2_score(y_test, y_predicted))
```

#### Here, predicted marks is around 93.7% as error rate is not abnormally large
